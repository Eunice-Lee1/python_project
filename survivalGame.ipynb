Project Description: The Survival Game's goal is to survive seven days without reaching a health of zero by doing a series of tasks necessary for survival. 

I have built a class called Player that creates the player of the game with the initial amount of resources and allows the player is allowed to choose their name. In order to make it out alive in the seven days, the player needs to choose carefully what tasks to do, depending on the resources he or she has and amount of health is remaining. Each function carries out the task needed to find food, eat food, find wood, and build a fire. Performing each task takes damage to whatever resource is being used and requires a certain amount of a resource to perform a task. For example, in order for the player to build a fire, he or she needs 5 points of wood, which builds a fire of two points. 

I have also built a class called SurvivalGame, which carries out the seven day challenge of survival. The player starts with a health of 50 points and knife durability of 7 points, which is needed to find food and wood. The player has 5 tasks to choose from and if the player doesn't have enough resources to do the task, a message should show indicating so. After each turn, a day passes by and health takes damage depending on what task is performed. 

```python
import os
#establish a class for creating the initial conditions of the player and necessary functions to do each task
class Player: 
    def __init__(self):
        self.name = ""
        self.wood = 0
        self.food = 0
        self.fire = 0
        self.knife_dur = 0
        self.health = 0
        
    def set_name(self,name):        
        self.name = name

    def build_fire(self):
        self.wood -= 5
        self.fire += 2

    def eat_food(self):
        self.food -= 1

    def find_food(self):
        if self.check_knife(1):
            self.knife_dur -= 1
            self.food += 1

    def find_wood(self):
        if self.check_knife(3):
            self.knife_dur -= 3
            self.wood += 5

    def check_knife(self, dmg):
        if self.knife_dur - dmg < 0:
            print("Your knife is as dull as my soul.")
            return False
        else:
            return True  

#establish a class to carry out the game, allowing player to choose a task based on available resources
class SurvivalGame:
    def __init__(self):
        self.player = Player()
        self.player.health = 50
        self.player.knife_dur = 7
        self.days = 7
        self.ate_food = False
    
    def start(self):
        name = input("What's your name?") 
        self.player.name = name
        output = ""
        
        #as long as the player's health is above 0 and there are days remaining, the player has a chance to live
        while self.player.health > 0 and self.days > 0:
            ate_food = False
            os.system('clear')
            self.print_player()
            print("\n" + output + "\n")
            output = ""
            print("What do you want to do?")
            choice = input("""1: Eat some food
2: Find some wood
3: Build a fire
4: Find some food 
5: Nothing (just die)""")
            if choice == '1':
                if self.player.food > 0:
                    self.player.eat_food()
                    output += "That was delicious"
                    self.ate_food = True
                else:
                    output += "You have no food! Sucks you can't eat."
                    
            elif choice == '2':
                    self.player.find_wood()
                    output += "Ya found some wood."
                    
            elif choice == '3':
                if self.player.wood > 4:
                    self.player.build_fire()
                    output += "I feel toasty"
                else: 
                    output += "You don't have wood. I guess you're just going to freeze to death."
                    
            elif choice == '4':
                self.player.find_food()
                output += "You're not going to starve."
            
            elif choice == '5':
                self.player.health = 0

            else:
            	output += "Can't you follow instructions, you wasted a day you fool."
                
            self.check_status(self.ate_food)
            self.new_day()
            self.endgame()
            
    
    def check_status(self, ate_food):      
        dmg = 0
        if self.player.fire > 0:
            self.player.fire -= 1
        else:
            dmg += 5
        
        if not ate_food:
            dmg += 5

        self.player.health -= dmg
        dmg = 0

    def new_day(self):      
        self.ate_food = False 
        self.days -= 1
    
    #the player survives if he or she makes it through 7 days with health greather than 0
    def endgame(self):
        if self.days == 0:
            print("Good job you made it alive")
        elif self.player.health <= 0:
        	print("YA DED")
        elif self.days > 0:
            print("Another day, another chance to live")
        else:
            print("Sorry you died")
    
    def print_player(self):
        print("""\
Status of {0}
health: {health}
wood: {1}
fire: {2}
food: {3}
knife durability: {4}
Day {5}""".format(self.player.name,
                  self.player.wood,
                  self.player.fire,
                  self.player.food,
                  self.player.knife_dur,
                  self.days,health = self.player.health))

game = SurvivalGame()
game.start()
```
